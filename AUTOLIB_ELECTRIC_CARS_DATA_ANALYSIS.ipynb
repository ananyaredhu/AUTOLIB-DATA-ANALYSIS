**DATA CLEANING**

1. VALIDITY

```
#removing unwanted columns
auto=auto.drop(['Displayed comment','Geo point','Scheduled at','Status','Address','Charging Status','Charge Slots','ID'],axis=1)
 
```

```
#merging year, month and day as date.
auto['dateint'] =auto['year'].astype(str) + auto['month'].astype(str).str.zfill(2) + auto['day'].astype(str).str.zfill(2)
auto['date']=pd.to_datetime(auto['dateint'])
```

```
# dropping the columns that are not reqiured
auto=auto.drop(['month','year','day','minute','dateint'],axis=1)
auto.head()
```

***DATA ANALYSIS***








1). The best  hour of the day for picking up a shared  electric car, in the city of Paris over month of April 2018.

```
# we filter our data to the month of April in the city of Paris and finding the number of cars available in the station,
#hence getting the most popular hour for picking a  shared car.
paris=auto[(auto['City']== 'Paris') & (auto['date']>='2018-04-01') & (auto['date']<='2018-04-30') & (auto['Rental status']=='operational')]
tables=pd.pivot_table(paris,values=['bluecar','utilib','utilib 1.4'],index=['hour'],aggfunc=np.sum)
tables
sum_cars = tables['bluecar'] +tables['utilib 1.4'] + tables['utilib']
tables['total_cars'] = sum_cars
tables.sort_values(by='total_cars',ascending=0).head(2)
#21hr and 6th hour are the best hours for picking up an electric car in Paris in the month of April
```

2). Most popular hour for returning cars

```

car_retur = pd.pivot_table(auto,values=['Slots'],index=['hour'],aggfunc=np.sum)
car_retur.sort_values(by='Slots',ascending=0).head(3)
#14th hour is the most popular hour for returning cars. Its the hour with the highest number of slots
```

3). What station is the most popular


```
# The station that is very popular.
#the popular station is the station with the highest number of cars available
auto['total_resources']=auto['bluecar']+auto['utilib']+auto['utilib 1.4']
auto.pivot_table(values=['total_resources'],index=['Public name'],aggfunc=np.sum).sort_values(by='total_resources',ascending=0).head(3)
```

```
# the most popular station At the most popular picking hour.
data=auto[auto['Rental status']=='operational']
data
pop=pd.pivot_table(data,values=['bluecar','utilib','utilib 1.4'],index=['hour'],aggfunc=np.sum)
pop
t=pop['bluecar']+pop['utilib'] +pop['utilib 1.4']
pop['total'] = t
pop
pop.sort_values(by='total',ascending=0).head(3)
#the 6th hour was the most popular picking hour.

station_hour=data[data['hour']==6]
public=pd.pivot_table(station_hour,values=['bluecar','utilib','utilib 1.4'],index=['Public name','Station type'],aggfunc=np.sum)
tot=public['bluecar']+public['utilib 1.4']+public['utilib']
public['total']=tot
public.sort_values(by='total',ascending=0).head(3)
#Garches/Raymond PoincarÃ©/102 is the most popular station at the popular picking hour
```
